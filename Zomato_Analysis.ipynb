```python
import pandas as pd
```

```python
df = pd.read_csv("zomato.csv", encoding='latin-1')

df.head(10)
```

```python
print(df.columns)
```

```python
print("Shape:", df.shape)
```

```python
df.info()
df.describe()
```

```python
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]
df.columns
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
df['cuisines'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

```python
print(df.columns.tolist())
```

```python
print("\nData types:")
print(df.dtypes)
```

```python
print("\nData overview:")
print(df.describe(include='all'))
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
print(missing[missing > 0])
```

```python
df.drop(['switch_to_order_menu', 'address', 'locality_verbose'], axis=1, inplace=True)
```

```python
df['has_table_booking'] = df['has_table_booking'].map({'Yes': 1, 'No': 0})
df['has_online_delivery'] = df['has_online_delivery'].map({'Yes': 1, 'No': 0})
df['is_delivering_now'] = df['is_delivering_now'].map({'Yes': 1, 'No': 0})
```

```python
print("Table Booking:")
print(df['has_table_booking'].value_counts())
print("\nOnline Delivery:")
print(df['has_online_delivery'].value_counts())
print("\nIs Delivering Now:")
print(df['is_delivering_now'].value_counts())
```

```python
print("Table Booking (0 = No, 1 = Yes):")
print(df['has_table_booking'].value_counts())

print("\nOnline Delivery (0 = No, 1 = Yes):")
print(df['has_online_delivery'].value_counts())

print("\nIs Delivering Now (0 = No, 1 = Yes):")
print(df['is_delivering_now'].value_counts())
```

```python
df['cuisines'].value_counts().head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['aggregate_rating'], bins=20, kde=True)
plt.title("Rating Distribution")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()
```

```python
sns.scatterplot(x='votes', y='aggregate_rating', data=df)
plt.title('Votes vs Rating')
plt.show()
```

```python
df['city'].value_counts().head(10).plot(kind='bar', figsize=(10,5), title='Top 10 Cities with Most Restaurants')
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
sns.boxplot(x='has_table_booking', y='aggregate_rating', data=df)
plt.title("Rating by Table Booking Availability")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
top_cities = df['city'].value_counts().head(10).index
avg_cost = df[df['city'].isin(top_cities)].groupby('city')['average_cost_for_two'].mean().sort_values()

avg_cost.plot(kind='barh', title='Average Cost for Two by Top Cities')
plt.xlabel("Average Cost")
plt.show()
```

```python
top_cuisines = df['cuisines'].value_counts().head(10).index
subset = df[df['cuisines'].isin(top_cuisines)]

sns.boxplot(x='cuisines', y='aggregate_rating', data=subset)
plt.xticks(rotation=45)
plt.title("Rating Distribution by Top Cuisines")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='votes', y='aggregate_rating', data=df, hue='price_range', palette='coolwarm')
plt.title('Votes vs Aggregate Rating (by Price Range)')
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.legend(title='Price Range')
plt.show()
```

```python
df.to_csv('cleaned_zomato_data.csv', index=False)
```
